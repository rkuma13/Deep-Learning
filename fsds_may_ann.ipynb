```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("/content/aps_data.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
np.nan
```

```python
dataset=dataset.replace("na",np.nan)
```

```python
pd.options.display.max_rows=1000
```

```python
dataset.isnull().sum()
#since we replaced the na to nan , now we can see the number of null values
```

```python
dataset.isnull().sum().sum()
```

```python
#create the dataframe for these null values
pd.DataFrame(dataset.isnull().sum().sort_values(ascending=False)).rename(columns={0:"Null Value Count"})
```

```python
### Filling null values in feature with median of that feature
for feature in [feature for feature in dataset.columns if feature not in ['class']]:
    dataset[feature]=dataset[feature].fillna(dataset[feature].median())
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
#segregate the data in X and Y
X= dataset.drop("class",axis =1)
```

```python
y=dataset["class"]
```

```python
X.info(verbose=True, show_counts=True)
```

```python
y=pd.get_dummies(y,drop_first=True)#we have converterd the data in numerical
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_test = np.asarray(X_test).astype(np.float32)
#Getting error for y_pred =classifier.predict(X_test) i.e,"y_pred =classifier.predict(X_test) ValueError: Failed to convert a NumPy array to a Tensor (Unsupported object type int)." SO I have used this code before model
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
sc.transform(X_test)
#we are able the scale the data
```

```python
# Build the Artificial Neural Network
```

```python
# Importing the Keras libraries and packages
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()

classifier.add(Dense(units = 10, kernel_initializer='he_uniform',activation = 'relu', input_dim = 170))#weight initializer is he_uniform and activation function is relu
classifier.add(Dense(units = 10, kernel_initializer='he_uniform',activation = 'relu'))
classifier.add(Dense(units = 1, kernel_initializer='glorot_uniform',activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.summary()
```

```python
model_history = classifier.fit(X_train,y_train, batch_size=10, epochs=20)
```

```python
model_history.history.keys()
```

```python
#components we are getting
model_history.history['accuracy']
```

```python
model_history = classifier.fit(X_train,y_train, batch_size=10, epochs=30,validation_split=0.25)
```

```python
#List all data in history
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model_accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#To save the model we use h5 
import time
import os

def saveModel_path(model_dir="SAVED_MODELS"):
    os.makedirs(model_dir, exist_ok=True)
    fileName = time.strftime("Model_%Y_%m_%d_%H_%M_%S_.h5")    
    model_path = os.path.join(model_dir, fileName)
    print(f"your model will be saved at the following location\n{model_path}")
    return model_path
```

```python
classifier.save(saveModel_path())
```

```python
from kerastuner.tuners import RandomSearch
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(units=32,kernel_initializer = 'he_uniform',activation='relu',input_dim=170))
  model.add(Dense(units = 32, kernel_initializer = 'he_uniform',activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  optimizer=hp.Choice("optimizer",values=["adam","sgd","rmsprop","adadelta"])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner = RandomSearch(build_model,objective='val_accuracy',max_trials=6,directory='project1',project_name='aps_fault')
```

```python
tuner.search(X_train,y_train,validation_data=(X_test,y_test),epochs=6)
```

```python
from tensorflow.keras import optimizers
def build_model(hp):
    model = Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(Dense(units=hp.Int('units_' + str(i),
                                            min_value=32,
                                            max_value=512,
                                            step=32),
                               activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy'])
    return model
     
```

```python
tuner_second=RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='project3',project_name='aps-fault-sensor')
```
